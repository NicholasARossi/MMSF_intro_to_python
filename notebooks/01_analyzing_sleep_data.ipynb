```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Loading Data

```python
sleep_df=pd.read_csv('../data/bb_sleep.csv')

sleep_df['Time']= pd.to_datetime(sleep_df['Time'])
sleep_df=sleep_df.set_index('Time')
```

# Plotting Data

```python
fig,ax=plt.subplots()
ax.plot(sleep_df['Duration(minutes)'])
plt.xticks(rotation=45)
```

```python
fig,ax=plt.subplots()
new_data=sleep_df['Duration(minutes)'].resample('D').sum()
ax.plot(new_data,color='black',linewidth=4)
plt.xticks(rotation=45)
```
