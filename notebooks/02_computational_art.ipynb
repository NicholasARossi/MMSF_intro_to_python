```python
import matplotlib.pyplot as plt
import numpy as np
```

# simple plot

```python
plt.close('all')
fig,ax=plt.subplots(figsize=(6,2))
x_vect=np.linspace(0,np.pi*6,1000)


#alpha= transparency
#s= marker size

ax.scatter(x_vect,np.sin(x_vect),alpha=0.03,s=abs(np.sin(x_vect))*100,color='black')
    
plt.axis('off')

fig.savefig('sin.png',dpi=300)
```

```python
plt.close('all')
fig,ax=plt.subplots(figsize=(6,2))
x_vect=np.linspace(0,np.pi*6,1000)

for l in np.linspace(0,4,100):
    ax.plot(x_vect+l,np.sin(x_vect),alpha=l/20,color='black')
    
plt.axis('off')

fig.savefig('sin2.png',dpi=300)
```

```python

plt.close('all')
fig,ax=plt.subplots(figsize=(4,4))
size=100
color_combos=[('yellow',1),('cyan',0.5),('magenta',0.5)]

for i,n in enumerate([3,4,5]):

    u=np.random.multivariate_normal([1,1],[[1, 0.5], [0.5, 1]],size=size)
    ax.scatter(u[:,0],u[:,1],color=color_combos[i][0],alpha=color_combos[i][1],s=80)
plt.axis('off')
```

# Advance Plot

```python
# initializing functions
def return_squiggle(i,n):
    theta=np.arange(0,1001)/1000
    r=np.cos(np.pi/n)/np.cos(2*np.pi*(n*theta)%1/n-np.pi/n)
    return r*np.cos(2*np.pi*theta),r*np.sin(2*np.pi*theta)
```

```python

plt.close('all')
fig,ax=plt.subplots(figsize=(4,4))
color_combos=[('yellow',1),('cyan',0.5),('magenta',0.5)]

for i,n in enumerate([3,4,5]):
    x,y=return_squiggle(i,n)
    ax.plot(x,y,linewidth=10,color=color_combos[i][0],alpha=color_combos[i][1])

plt.axis('off')
fig.savefig('circle.png',dpi=300)
```

```python
plt.close('all')
fig,ax=plt.subplots(figsize=(4,4))


for i,n in enumerate(np.arange(3,40)):
    x,y=return_squiggle(i,n)
    ax.plot(x,y,linewidth=1,color='black',alpha=0.1)

plt.axis('off')
fig.savefig('circle2.png',dpi=300)
```
